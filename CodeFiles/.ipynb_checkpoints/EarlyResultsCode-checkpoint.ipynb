```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
```

## Event study graph

```python
df = pd.read_csv("../Data/final_merged_panel.csv").copy()
df = df.sort_values(["metro_id", "econ_year"]).reset_index(drop=True)

# outcome: log metro GDP
df["log_gdp"] = np.log(df["metro_gdp"])

#first observeed super bowl win
first_treat = (
    df.loc[df["any_super_bowl_win"] == 1]
      .groupby("metro_id")["econ_year"]
      .min()
)

df["treatment_time"] = df["metro_id"].map(first_treat)
df["ever_treated"] = df["treatment_time"].notna().astype(int)
df["event_time"] = df["econ_year"] - df["treatment_time"]

window = range(-3, 4)
for k in window:
    if k == -1:
        continue
    varname = f"event_{k}".replace("-", "m")
    df[varname] = ((df["event_time"] == k) & (df["ever_treated"] == 1)).astype(int)

event_vars = [f"event_{k}".replace("-", "m") for k in window if k != -1]

#TWFE study
formula = "log_gdp ~ C(metro_id) + C(econ_year) + " + " + ".join(event_vars)

model = smf.ols(formula, data=df).fit(
    cov_type="cluster",
    cov_kwds={"groups": df["metro_id"]}
)

#collect the coefficents
rows = []
for k in window:
    if k == -1:
        continue
    varname = f"event_{k}".replace("-", "m")
    coef = model.params.get(varname, np.nan)
    se = model.bse.get(varname, np.nan)

    rows.append({
        "event_time": k,
        "coef": coef,
        "se": se,
        "lower_95": coef - 1.96 * se,
        "upper_95": coef + 1.96 * se
    })

es = pd.DataFrame(rows).sort_values("event_time")

# add omitted baseline point at -1
baseline = pd.DataFrame({
    "event_time": [-1],
    "coef": [0.0],
    "se": [0.0],
    "lower_95": [0.0],
    "upper_95": [0.0]
})

es_plot = pd.concat([es, baseline], ignore_index=True).sort_values("event_time")

# verifying the output
print("\nFIRST TREATMENT YEAR BY METRO")
print(first_treat.sort_values())

print("\nNUMBER OF EVER-TREATED METROS")
print(df[["metro_id", "ever_treated"]].drop_duplicates()["ever_treated"].value_counts())

print("\nEVENT-TIME SUPPORT")
print(
    df.loc[df["ever_treated"] == 1, "event_time"]
      .value_counts()
      .sort_index()
      .loc[lambda s: (s.index >= -3) & (s.index <= 3)]
)

print("\nEVENT STUDY COEFFICIENTS")
print(es.round(4))

# coverting logs in percentage so it is easier to interpret
es["pct_effect"] = 100 * (np.exp(es["coef"]) - 1)
es["pct_lower"] = 100 * (np.exp(es["lower_95"]) - 1)
es["pct_upper"] = 100 * (np.exp(es["upper_95"]) - 1)

print("\nAPPROXIMATE PERCENT EFFECTS")
print(es[["event_time", "pct_effect", "pct_lower", "pct_upper"]].round(2))

# ending ploot
plt.figure(figsize=(10, 6))
plt.axhline(0, color="black", linestyle="--", linewidth=1)
plt.axvline(-0.5, color="red", linewidth=1)

plt.errorbar(
    es_plot["event_time"],
    es_plot["coef"],
    yerr=[
        es_plot["coef"] - es_plot["lower_95"],
        es_plot["upper_95"] - es_plot["coef"]
    ],
    fmt="o",
    capsize=4
)

plt.title("Event Study: First Observed Super Bowl Win")
plt.xlabel("Event Time (years relative to first observed win)")
plt.ylabel("Effect on log metro GDP")
plt.xticks(list(window))
plt.tight_layout()
plt.show()
```

## ATT Plot

```python
#definign first super bowl win
first_treat = (
    df.loc[df["any_super_bowl_win"] == 1]
      .groupby("metro_id")["econ_year"]
      .min()
)

df["treatment_time"] = df["metro_id"].map(first_treat)
df["ever_treated"] = df["treatment_time"].notna().astype(int)

# what is considered treated
df["treated"] = (
    (df["ever_treated"] == 1) &
    (df["econ_year"] >= df["treatment_time"])
).astype(int)

# Construct the outcomes
df["log_gdp"] = np.log(df["metro_gdp"])
df["log_income_pc"] = np.log(df["metro_personal_income_pc"])

# growth = within-metro log difference
df["gdp_growth"] = df.groupby("metro_id")["log_gdp"].diff()
df["income_growth"] = df.groupby("metro_id")["log_income_pc"].diff()

# helper function to estimate the ATT, recommended online
def estimate_att(data, outcome, label):
    reg_df = data[[outcome, "treated", "metro_id", "econ_year"]].dropna().copy()

    formula = f"{outcome} ~ treated + C(metro_id) + C(econ_year)"
    model = smf.ols(formula, data=reg_df).fit(
        cov_type="cluster",
        cov_kwds={"groups": reg_df["metro_id"]}
    )

    coef = model.params["treated"]
    se = model.bse["treated"]

    return {
        "specification": label,
        "outcome": outcome,
        "att": coef,
        "se": se,
        "lower_95": coef - 1.96 * se,
        "upper_95": coef + 1.96 * se,
        "n_obs": len(reg_df),
        "n_metros": reg_df["metro_id"].nunique(),
        "model": model
    }

# estimate the ATT
results = []
results.append(estimate_att(df, "log_gdp", "Log Metro GDP"))
results.append(estimate_att(df, "gdp_growth", "Metro GDP Growth"))
results.append(estimate_att(df, "log_income_pc", "Log Income per Capita"))
results.append(estimate_att(df, "income_growth", "Income per Capita Growth"))

att_df = pd.DataFrame(results)

# approximate percent effects for the log-level outcomes
att_df["pct_effect"] = np.nan
att_df["pct_lower"] = np.nan
att_df["pct_upper"] = np.nan

level_mask = att_df["outcome"].isin(["log_gdp", "log_income_pc"])
att_df.loc[level_mask, "pct_effect"] = 100 * (np.exp(att_df.loc[level_mask, "att"]) - 1)
att_df.loc[level_mask, "pct_lower"] = 100 * (np.exp(att_df.loc[level_mask, "lower_95"]) - 1)
att_df.loc[level_mask, "pct_upper"] = 100 * (np.exp(att_df.loc[level_mask, "upper_95"]) - 1)

# print verification output
print("\nFIRST TREATMENT YEAR BY METRO")
print(first_treat.sort_values())

print("\nNUMBER OF EVER-TREATED METROS")
print(df[["metro_id", "ever_treated"]].drop_duplicates()["ever_treated"].value_counts())

print("\nATT RESULTS")
print(
    att_df[[
        "specification", "att", "se", "lower_95", "upper_95", "n_obs", "n_metros"
    ]].round(4)
)

print("\nAPPROXIMATE PERCENT EFFECTS FOR LOG-LEVEL OUTCOMES")
print(
    att_df.loc[level_mask, [
        "specification", "pct_effect", "pct_lower", "pct_upper"
    ]].round(2)
)

# optional: print treated coefficient table from each model
for row in results:
    print(f"\n--- {row['specification']} ---")
    print(row["model"].summary().tables[1])

# plot
plot_df = att_df.copy()
plot_df["x"] = np.arange(len(plot_df))

plt.figure(figsize=(10, 6))
plt.axhline(0, color="black", linestyle="--", linewidth=1)

plt.errorbar(
    plot_df["x"],
    plot_df["att"],
    yerr=[
        plot_df["att"] - plot_df["lower_95"],
        plot_df["upper_95"] - plot_df["att"]
    ],
    fmt="o",
    capsize=5
)

plt.xticks(plot_df["x"], plot_df["specification"], rotation=15)
plt.ylabel("Estimated ATT")
plt.title("Average Treatment Effects of First Observed Super Bowl Wins")
plt.tight_layout()
plt.show()
```
